```python
import newspaper
import pandas as pd
```

## Arabian news parser

```python
arab_paper = newspaper.build('https://www.skynewsarabia.com/', language='ar')
```

```python
for category in arab_paper.category_urls():
  print(category)
```

```python
len(arab_paper.articles)
```

```python
news_text, news_article = [], []
k = 0
for i in arab_paper.articles:
  if k == 50:
    break
  else:
    article = i
    try:
      article.download()
      article.parse()
      if article.text != '':
        news_article.append(article.title)
        news_text.append(article.text)
        k += 1
    except:
      pass
```

```python
arab_parsed = pd.DataFrame([news_article, news_text]).T
arab_parsed.columns = ['article', 'text']
arab_parsed['language'] = 'ar'
arab_parsed.head()
```

## All data

```python
final_data = pd.concat([arab_parsed, us_parsed, france_parsed]).reset_index().drop(columns=['index'])

final_data
```

## Clean data

```python
import re

# clean text from html tags and /n
def cleantext(raw_text):
  html_clean = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
  n_clean = re.compile('\n')
  cleantext = re.sub(html_clean, '', raw_text)
  cleantext = re.sub(n_clean, ' ', cleantext)
  cleantext = cleantext.replace('*', '')
  cleantext = re.sub('  ', ' ', cleantext)
  return cleantext
```

```python
for i in range(len(final_data)):
    final_data['article'][i] = cleantext(final_data['article'][i])
    final_data['text'][i] = cleantext(final_data['text'][i])
    
final_data
```

```python
final_data.to_excel('news_data.xlsx', index=False)
```
